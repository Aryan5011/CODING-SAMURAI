```python
property = data['room_type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(property, labels= property.index, autopct='%1.1f%%', startangle=180)
plt.title('Distribution of Property Types')
plt.show()
```

```python
property = data['host_response_time'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(property, labels=property.index, autopct='%1.1f%%', startangle=180)
plt.title('Distribution of Property Types')
plt.show()
```

```python
property = data['bed_type'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(property, labels=property.index, autopct='%1.1f%%', startangle=180)
plt.title('Distribution of Bed Types')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='property_type')
plt.title('Distribution of Property Types')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(data=data, x='neighbourhood')
plt.title('Distribution of Neighbourhoods')
plt.xlabel('Neighbourhood')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=data, x='instant_bookable')
plt.title('Count of Instant Bookable Listings')
plt.xlabel('Instant Bookable')
plt.ylabel('Count')
plt.show()
```

```python
data = data.drop('scrape_id', axis=1)
```

```python
numeric_columns = data.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(20,40))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap for Airbnb Dataset')
plt.show()
```
